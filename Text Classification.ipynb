```python
dictionary = {}
```

```python
# Building the vocabulary
for current_path in training_paths:
    f = open(current_path[0])
    for i in f.read().split('\n'):
        line = i.split(' ')
        for j in line:
            if j.lower() not in stopwords:
                if j.lower() not in dictionary:
                    dictionary[j.lower()] = 1
                else:
                    dictionary[j.lower()] = dictionary[j.lower()] + 1
```

```python
# sorting dictionary
dictionary_sorted = sorted(dictionary.items(), key = lambda x : x[1])
len(dictionary_sorted)
```

```python
# Keeping only top 10000 words
del dictionary_sorted[:337426]
```

```python
len(dictionary_sorted)
```

```python
# words 
words = {}
j = 0
for i in dictionary_sorted:
    words[i[0]] = j
    j = j + 1
```

```python
# Function will return 2D list of X and 1D list of Y  
def build_data(paths):
    X = []
    Y = []
    for current_path in paths:
        # opening the current_path file and storing frequency of words in X_current and then append it to X
        f = open(current_path[0])
        X_current = [0 for i in range(10000)]
        for i in f.read().split('\n'):
            line = i.split(' ')
            for j in line:
                word = j.lower()
                if word in words:
                    # words[word] will give the column no. of word
                    X_current[words[word]] = X_current[words[word]] + 1
        X.append(X_current)
        Y.append(current_path[1])
    return X,Y
```

```python
X_train,Y_train = build_data(training_paths)
```

```python
X_test,Y_test = build_data(testing_paths)
```

# Applying own implementation of Multinomial Naive Bayes

```python
def fit(X_train, Y_train):
    result = {}
    classes, counts = np.unique(Y_train, return_counts=True)
    for i in range(len(classes)):
        curr_class = classes[i]
        result["TOTAL_DATA"] = len(Y_train)
        result[curr_class] = {}
        X_tr_curr = X_train[Y_train == curr_class]
        s = 0
        for j in words:
            result[curr_class][words[j]] = X_tr_curr[:,words[j]].sum() 
            s = s + result[curr_class][words[j]]
        result[curr_class]["TotalWords"] = s       
        result[curr_class]["TOTAL_COUNT"] = counts[i]
    return result
```

```python
def log_probablity(dictionary_train, x, curr_class):
    output = np.log(dictionary_train[curr_class]["TOTAL_COUNT"]) - np.log(dictionary_train["TOTAL_DATA"])
    for j in range(10000):
        if x[j] != 0:
            num = x[j]*(dictionary_train[curr_class][j]+1)
            deno = dictionary_train[curr_class]["TotalWords"]+10000
            p = np.log(num) - np.log(deno)
            output = output + p
    return output
```

```python
def predictSinglePoint(dictionary_train, x):
    classes = dictionary_train.keys()
    best_p = -100000000
    best_class = -10000000
    for curr_class in classes:
        if(curr_class == "TOTAL_DATA"):
            continue
        p_curr_class = log_probablity(dictionary_train, x, curr_class)
        if(p_curr_class > best_p):
            best_p = p_curr_class
            best_class = curr_class
    return best_class
```

```python
def predict(dictionary_train, X_test):
    Y_pred = []
    n = X_test.shape[0]
    for i in range(n):
        y_predicted = predictSinglePoint(dictionary_train, X_test[i])
        Y_pred.append(y_predicted)
    return Y_pred
```

```python
train_dictionary = fit(X_tr, Y_tr)
```

```python
Y_pred_2 = predict(train_dictionary, X_te)
```

```python
print(accuracy_score(Y_te,Y_pred_2))
print(classification_report(Y_te, Y_pred_2))
```
